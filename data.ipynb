```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hitori_metrics.csv')
df.head()
```

```python
df['Time_group'] = df['Time'].apply(lambda x: round(x, 1))  # Round to nearest hundredth for example
grouped = df.groupby(['FunctionCalls', 'Time_group'])
filtered_df = grouped.filter(lambda x: len(x) > 1)
```

```python
print(filtered_df)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['FunctionCalls'], df['Time'])
plt.title('Function Calls vs Time')
plt.xlabel('Function Calls')
plt.ylabel('Time')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Difficulty', y='FunctionCalls', data=df)
plt.title('Boxplot of Function Calls for each Difficulty Level')
plt.xlabel('Difficulty')
plt.ylabel('Function Calls')
plt.show()
```
